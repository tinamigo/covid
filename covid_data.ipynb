```python
import matplotlib.pylab as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('data2.csv')
```

```python
populosos = df[df.population > 3000000] #poblacion mayor a 3 millones
```

```python
pop_sin_outliers = df[df.gdp_per_capita < 70000]
```

```python
sns.scatterplot(data=df, y="people_vaccinated_per_hundred", x="gdp_per_capita")
plt.text(18933.907,28.86,"Arg.")
plt.show()
```

```python
sns.lmplot( y="people_vaccinated_per_hundred", x="gdp_per_capita", data=df)
plt.text(18933.907,28.86,"Arg.")
plt.show()
```

```python
sns.lmplot(y="people_vaccinated_per_hundred", x="gdp_per_capita", data=populosos)
plt.text(18933.907,28.86,"Arg.")
plt.show()
```

```python
sns.lmplot(y="people_vaccinated_per_hundred", x="gdp_per_capita", data=pop_sin_outliers)
plt.text(18933.907,28.86,"Arg.")
plt.show()
```
