```python
import matplotlib.pylab as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('data2.csv')
```

```python
populosos = df[df.population > 3000000]
```

```python
pop_sin_outliers = df[df.people_vaccinated_per_hundred < 30]
pop_sin_outliers = pop_sin_outliers[pop_sin_outliers.gdp_per_capita < 55000]
pop_sin_outliers
```

```python
sns.scatterplot(data=df, x="people_vaccinated_per_hundred", y="gdp_per_capita")
plt.text(28.86,18933.907,"Arg.")
plt.show()
```

```python
sns.lmplot(x="people_vaccinated_per_hundred", y="gdp_per_capita", data=df)
plt.text(28.86,18933.907,"Arg.")
plt.show()
```

```python
sns.lmplot(x="people_vaccinated_per_hundred", y="gdp_per_capita", data=populosos)
plt.text(28.86,18933.907,"Arg.")
plt.show()
```

```python
sns.lmplot(x="people_vaccinated_per_hundred", y="gdp_per_capita", data=pop_sin_outliers)
plt.text(28.86,18933.907,"Arg.")
plt.show()
```
